```python
import matplotlib.pyplot as plt
y = [3,11,3,0]
x = ["beef","cheese","steak","chicken"]
plt.bar(x,y,)
plt.xlabel('Type of Pies')
plt.ylabel('Stock')
plt.title('Pie Bar Graph')
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
 
style.use('ggplot')
x = ["nik niks","simba","lays","fritos"]
y = [8,9,7,4]
plt.plot(x,y,'y',label='line graph', linewidth=5)
plt.title('Chips category')
plt.ylabel('Stock')
plt.xlabel('type of chips')
plt.legend()
plt.grid(True,color='k')
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
 
style.use('ggplot')
x = ["Mint","Caramel","Vanilla","Nuts"]
y = [8,12,5,9]
plt.plot(x,y,'g',label='line graph', linewidth=5)
plt.title('Chocolate category')
plt.ylabel('Stock')
plt.xlabel('Type of chocolate')
plt.legend()
plt.grid(True,color='k')
plt.show()
```

```python
y = ["apple","banana","orange","peach"]
x = [6,8,12,9]
# Plot list 'y'

plt.plot(y,x)

# Plot axes labels and show the plot

plt.xlabel('Type of Fruit', fontsize= 11)
plt.title('fruit Line Graph')
plt.ylabel('Stock', fontsize= 11)

plt.show()
```

```python
x = ["Onion","Tomato","Cabbage","Betroot"]
y = [17,14,15,9]
plt.plot(y, 'g')
plt.plot(y, 'bo')
plt.plot(x,y)
# Plot axes labels and show the plot

plt.xlabel('Types of Veggies')
plt.title('Veggies')
plt.ylabel('stock')

plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
x = ["plain","vanilla","choc","red velvet"]
y = [8,13,16,0]

plt.plot(x,y, color='purple')

# Customizing the axes

plt.xticks(rotation=45, color='teal', size=12)

plt.yticks(rotation=45, color='teal', size=12)

# Setting axes labels
plt.title( 'Cupcake Graph',{'color': 'orange', 'fontsize':15})

plt.xlabel('Types of Cupcakes', {'color': 'orange', 'fontsize':13})

plt.ylabel('Stock', {'color': 'orange', 'fontsize':13})

plt.show()
plt.savefig('books_read.jpg')
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = [12,12,12,12]

y = [8,12,5,9]

color = np.random.rand(4)

size = np.random.randint(50, 100, 10)

plt.scatter(x, y, c=color, s=size)

# Labeling the plot and showing it

plt.xlabel('Values from list x')

plt.ylabel('Values from list y')
plt.title('price vs stock',{'fontsize':15})
plt.show()
plt.savefig('books_read.png')
```
